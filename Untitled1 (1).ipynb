The Human Footprint (HFI) measures the impact of human activities on the environment by quantifying the pressure exerted on ecological processes and natural landscapes. This index highlights growing global concerns over biodiversity loss and ecological sustainability. Based on Sanderson's (2002) framework, the Human Footprint Index integrates datasets representing various anthropogenic pressures:

Built-up Index
Population Density
Nighttime Lights
Road Networks (specific to the U.S.)
Cropland
Pastureland
This workflow enables users to compute the Human Footprint Index (HFI) for any specified coordinates within the United States, spanning the years 2000 to 2020.

Step 1: Defines functions necessary for HFI calculation.

```python
!pip install rasterio
import geemap
from google.colab import drive
import ee
import rasterio
from rasterio.warp import calculate_default_transform, reproject, Resampling
import os
import numpy as np
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt


ee.Authenticate()
ee.Initialize(project='ee-my-usernameurbannnnn')


drive.mount('/content/drive')

#Check metadata of exported tiff file
def check_tiff_metadata(tiff_path):
    with rasterio.open(tiff_path) as dataset:
        crs = dataset.crs
        resolution = dataset.res
        bounds = dataset.bounds

        metadata = {
            "File": tiff_path,
            "CRS": crs.to_string(),
            "Resolution": resolution,
            "Bounds": bounds
        }
    return metadata

#function to loop through all the tiff files in the folder to check metadata
def loop_through_tiffs(directory):
    tiff_metadata_list = []

    for filename in os.listdir(directory):
        if filename.endswith(".tif") or filename.endswith(".tiff"):
            tiff_path = os.path.join(directory, filename)
            metadata = check_tiff_metadata(tiff_path)
            tiff_metadata_list.append(metadata)
            print(f"Processed metadata for: {filename}")

    return tiff_metadata_list

#function to reproject raster to a target crs

def custom_reproject_raster(input_path, output_path, target_crs, target_transform, target_shape, resampling_method=Resampling.bilinear):
    with rasterio.open(input_path) as src:
        # read source data
        src_data = src.read(1)
        src_profile = src.profile

        # target raster profile
        src_profile.update({
            "crs": target_crs,
            "transform": target_transform,
            "height": target_shape[0],
            "width": target_shape[1],
        })

        # reproject the raster
        with rasterio.open(output_path, "w", **src_profile) as dst:
            reproject(
                source=src_data,
                destination=rasterio.band(dst, 1),
                src_transform=src.transform,
                src_crs=src.crs,
                dst_transform=target_transform,
                dst_crs=target_crs,
                resampling=resampling_method
            )
    print(f"Reprojected: {input_path} -> {output_path}")



def batch_reproject_to_target(directory, target_raster, output_directory):

    with rasterio.open(target_raster) as target_ds:
        target_crs = target_ds.crs
        target_transform = target_ds.transform
        target_shape = (target_ds.height, target_ds.width)

    os.makedirs(output_directory, exist_ok=True)

    # loop through all TIFF files
    for filename in os.listdir(directory):
        if filename.lower().endswith(".tif") or filename.lower().endswith(".tiff"):
            input_path = os.path.join(directory, filename)
            output_path = os.path.join(output_directory, filename)

            try:

                custom_reproject_raster(
                    input_path=input_path,
                    output_path=output_path,
                    target_crs=target_crs,
                    target_transform=target_transform,
                    target_shape=target_shape,
                    resampling_method=Resampling.bilinear
                )
            except TypeError as e:
                print(f"TypeError reprojecting {input_path}: {str(e)}")
            except Exception as e:
                print(f"Error reprojecting {input_path}: {str(e)}")

# Function to define AOI
def define_aoi():
    print("Define your Area of Interest (AOI) by entering four corner coordinates.")
    print("Enter coordinates in the format: longitude latitude (e.g., 89.0846 35.6904).")
    aoi_coordinates = []
    for i in range(4):
        coord = input(f"Enter coordinate {i + 1}: ")
        lon, lat = map(float, coord.split())
        aoi_coordinates.append([lon, lat])
    aoi_coordinates.append(aoi_coordinates[0])
    print("AOI defined.")
    return ee.Geometry.Polygon(aoi_coordinates)

# Function to display dataset options and get the user's choice
def select_dataset(processed_datasets):
    datasets = {
        1: "COPERNICUS/S2_SR_HARMONIZED",
        2: "CIESIN/GPWv411/GPW_Population_Density",
        3: "NOAA/VIIRS/DNB/ANNUAL_V22",
        4: "TIGER/2016/Roads",
        5: "USGS/NLCD_RELEASES/2021_REL/NLCD"
    }
    print("\nChoose a dataset to process:")
    for key, value in datasets.items():
        if key not in processed_datasets:
            print(f"{key}. {value}")
    choice = int(input("Enter the number corresponding to your choice: "))
    if choice in datasets and choice not in processed_datasets:
        return choice, datasets[choice]
    else:
        print("Invalid choice or dataset already processed.")
        return None, None

# Function to get the specific year for a dataset from the user
def get_year_for_dataset(dataset):
    if dataset == "CIESIN/GPWv411/GPW_Population_Density":
        print("Choose a year:")
        print("1. 2000")
        print("2. 2005")
        print("3. 2010")
        print("4. 2015")
        print("5. 2020")
        year_options = {1: "2000", 2: "2005", 3: "2010", 4: "2015", 5: "2020"}
        year_choice = int(input("Enter the number corresponding to your choice: "))
        return year_options.get(year_choice, None)
    elif dataset == "NOAA/VIIRS/DNB/ANNUAL_V22":
        print("Choose a year:")
        print("1. 2022")
        print("2. 2023")
        year_options = {1: "2022", 2: "2023"}
        year_choice = int(input("Enter the number corresponding to your choice: "))
        return year_options.get(year_choice, None)
    elif dataset == "TIGER/2016/Roads":
        print("Only the year 2016 is available for this dataset.")
        return "2016"
    elif dataset.startswith("USGS/NLCD_RELEASES/2021_REL/NLCD"):
        print("This script uses the latest NLCD release only (2021).")
        return "2021"
    else:
        return None


# Function to filter image collections
def filter_collection(dataset, aoi, year=None):
    if dataset == "COPERNICUS/S2_SR_HARMONIZED":
        start_date = input("Enter the start date (YYYY-MM-DD): ")
        end_date = input("Enter the end date (YYYY-MM-DD): ")
        cloud_cover = int(input("Enter the maximum cloud cover percentage: "))
        return ee.ImageCollection(dataset).filterBounds(aoi).filterDate(start_date, end_date).filter(ee.Filter.lt("CLOUDY_PIXEL_PERCENTAGE", cloud_cover))
    elif dataset in ["TIGER/2016/Roads", "USGS/NLCD_RELEASES/2021_REL/NLCD"]:
        return ee.ImageCollection(dataset).filterBounds(aoi)
    elif year:
        return ee.ImageCollection(dataset).filterDate(f"{year}-01-01", f"{year}-12-31").filterBounds(aoi)
    else:
        raise ValueError(f"A year must be provided for dataset '{dataset}'.")



# Function to process datasets and visualize
def process_and_visualize(dataset, collection, aoi, map_object, result_layers, year=None):
    if dataset == "COPERNICUS/S2_SR_HARMONIZED":
        median_image = collection.median().clip(aoi)
        B4 = median_image.select('B4')
        B8 = median_image.select('B8')
        B11 = median_image.select('B11')
        ndvi = B8.subtract(B4).divide(B8.add(B4)).rename("NDVI")
        ndbi = B11.subtract(B8).divide(B11.add(B8)).rename("NDBI")
        BU = ndvi.subtract(ndbi).rename("BU")
        vis_params = {"min": -1, "max": 1, "palette": ['blue', 'white', 'red']}
        map_object.addLayer(BU, vis_params, "Built-Up Index (BU)")
        result_layers.append((BU, "BU"))
    elif dataset == "CIESIN/GPWv411/GPW_Population_Density":
        total_density = collection.sum().clip(aoi).rename("Population Density")
        vis_params = {"min": 0, "max": 1000, "palette": ['blue', 'green', 'yellow', 'orange', 'red']}
        map_object.addLayer(total_density, vis_params, "Population Density")
        result_layers.append((total_density, "Population Density"))
    elif dataset == "NOAA/VIIRS/DNB/ANNUAL_V22":
        ntl_collection = ee.ImageCollection(dataset) \
            .filterDate(f"{year}-01-01", f"{year}-12-31") \
            .filterBounds(aoi) \
            .select('median')
        ntl_image = ntl_collection.median().clip(aoi).rename("Nighttime Lights")
        vis_params = {"min": 0, "max": 100, "palette": ['black', 'grey', 'white']}
        map_object.addLayer(ntl_image, vis_params, "Nighttime Lights")
        result_layers.append((ntl_image, "Nighttime Lights"))
    elif dataset == "TIGER/2016/Roads":
        roads = ee.FeatureCollection('TIGER/2016/Roads').map(lambda f: f.set('value', 1))
        roads_raster = roads.reduceToImage(properties=['value'], reducer=ee.Reducer.first()).unmask(0).clip(aoi).rename("Roads")
        vis_params = {"min": 0, "max": 1, "palette": ['white', 'blue']}
        map_object.addLayer(roads_raster, vis_params, "Roads")
        result_layers.append((roads_raster, "Roads"))
    elif dataset == "USGS/NLCD_RELEASES/2021_REL/NLCD":
        nlcd2021 = collection.first()
        landcover = nlcd2021.select('landcover')

        cropland = landcover.eq(82).multiply(7).clip(aoi).rename("Cropland")
        vis_params_cropland = {"min": 0, "max": 7, "palette": ['white', 'green']}
        map_object.addLayer(cropland, vis_params_cropland, "Cropland")
        result_layers.append((cropland, "Cropland"))

        pasture = landcover.eq(81).multiply(4).clip(aoi).rename("Pasture")
        vis_params_pasture = {"min": 0, "max": 4, "palette": ['white', 'blue']}
        map_object.addLayer(pasture, vis_params_pasture, "Pasture")
        result_layers.append((pasture, "Pasture"))

    map_object.addLayer(aoi, {"color": "yellow"}, "AOI Boundary")
    map_object.addLayerControl()

# Data calling and processingscript
aoi = define_aoi()
processed_datasets = []
result_layers = []
Map = geemap.Map(center=[40, -100], zoom=4)

while len(processed_datasets) < 5:
    choice, dataset = select_dataset(processed_datasets)
    if dataset:
        year = None

        if dataset in ["CIESIN/GPWv411/GPW_Population_Density", "NOAA/VIIRS/DNB/ANNUAL_V22"]:
            year = get_year_for_dataset(dataset)
            if not year:
                print("Invalid year selection. Please try again.")
                continue

        collection = filter_collection(dataset, aoi, year=year)
        process_and_visualize(dataset, collection, aoi, Map, result_layers, year=year)
        processed_datasets.append(choice)
        print(f"Dataset '{dataset}' processed and added to the map.")

Map

```

Step 2- EXport the processed results to google drive.

```python


# prompt user for Earth Engine project name
project_name = input("Next step is to export the result rasters to Google Drive.\nProvide Earth Engine Project name,without'': ")


ee.Authenticate()
ee.Initialize(project=project_name)


drive.mount('/content/drive')

# define raster names and corresponding outputs
raster_names = [
    ("BU", "BU_Raster"),
    ("Population Density", "PD_Raster"),
    ("Nighttime Lights", "NTL_Raster"),
    ("Roads", "Roads_Raster"),
    ("Cropland", "CROP_Raster"),
    ("Pasture", "PASTURE_Raster")
]

# export folder
folder_name = "GEE_Exports"

# loop through raster names and export each raster
for raster_name, image_name in raster_names:
    try:

        raster = next((r for r, name in result_layers if name == raster_name), None)

        if raster is None:
            print(f"Warning: Raster '{raster_name}' not found in result_layers. Skipping...")
            continue


        region = aoi.bounds().getInfo()['coordinates']


        task = ee.batch.Export.image.toDrive(
            image=raster,
            description=f"{image_name}_Export",
            folder=folder_name,
            fileNamePrefix=image_name,
            region=region,
            fileFormat='GeoTIFF',
            maxPixels=1e13
        )
        task.start()
        print(f"Export task for '{image_name}' started. Check Google Drive folder '{folder_name}' for output.")
    except Exception as e:
        print(f"Error exporting raster '{raster_name}': {str(e)}")
```

Step 4- Reproject and resample the exported tiff files into projected coordinate system.

```python

# paths
input_directory = "/content/drive/MyDrive/GEE_Exports"
target_raster = "/content/drive/MyDrive/GEE_Exports/CROP_Raster (1).tif"
output_directory = "/content/drive/MyDrive/GEE_Exports/Reprojected_Rasters"

# user message
print("Next step is to reproject the TIFFs into a projected coordinate system.")
print("The TIFFs will be reprojected into a projected coordinate system, in this case the CRS of NLCD is chosen as the CRS.")
print("Reprojecting exported TIFFs to the NLCD CRS is a necessary step and cannot be skipped.")


try:
    batch_reproject_to_target(input_directory, target_raster, output_directory)
    print(f"Reprojection complete. Reprojected rasters are saved in: {output_directory}")
except Exception as e:
    print(f"Error during reprojection: {str(e)}")
```

Step 6- Compute HFI

```python
#Compute HFI
# Paths
input_directory = "/content/drive/MyDrive/GEE_Exports/Reprojected_Rasters"
output_file = "/content/drive/MyDrive/GEE_Exports/Reprojected_Rasters/HFI_Raster.tif"

input_files = [
    "reprojected_BU_Raster (1).tif",
    "reprojected_NTL_Raster (1).tif",
    "reprojected_PD_Raster (1).tif",
    "reprojected_Roads_Raster (1).tif",
    "reprojected_CROP_Raster (1).tif",
    "reprojected_PASTURE_Raster (1).tif"
]
calculate_hfi = input("\nDo you want to compute the Human Footprint Index (HFI)? (yes/no): ")

if calculate_hfi in ['yes', 'y']:
    try:

        print("Calculating the Human Footprint Index (HFI)...")
        #opens, reads and converts the first band of the raster into 32-bit floating-point
        with rasterio.open(os.path.join(input_directory, input_files[0])) as src:
            hfi = src.read(1).astype(np.float32)
            profile = src.profile

        for file_name in input_files[1:]:
            file_path = os.path.join(input_directory, file_name)
            with rasterio.open(file_path) as src:
                hfi += src.read(1).astype(np.float32)

        # write the HFI raster
        with rasterio.open(output_file, 'w', **profile) as dst:
            dst.write(hfi, 1)

        print(f"HFI raster created and saved at {output_file}.")
    except Exception as e:
        print(f"Error calculating HFI: {str(e)}")
else:
    print("HFI calculation skipped.")




```

Step 7- Visualize HFI

```python


# user message to visualize HFI
visualize_hfi = input("\nDo you want to visualize the Human Footprint Index (HFI)? (yes/no): ")

if visualize_hfi in ['yes', 'y']:
    try:

        palette = [
            "#543005", "#8c510a", "#bf812d", "#dfc27d",
            "#f6e8c3", "#f5f5f5", "#c7eae5", "#80cdc1",
            "#35978f", "#01665e", "#003c30"
        ]


        custom_cmap = mcolors.LinearSegmentedColormap.from_list("CustomPalette", palette)


        plt.figure(figsize=(10, 8))
        raster_plot = plt.imshow(hfi, cmap=custom_cmap)
        plt.colorbar(raster_plot, label="HFI Value")
        plt.title("Human Footprint Index (HFI)")
        plt.axis("off")
        plt.show()

    except Exception as e:
        print(f"Error visualizing HFI: {str(e)}")
else:
    print("Visualization skipped.")
```
